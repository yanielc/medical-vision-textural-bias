```python
def summary_df(path:str, metric:str):
    '''Returns dataframe with evaluations of models
    on the requested class metric'''
    
    metrics_map ={'mean':0, 'et':1, 'tc':2, 'wt':3}

    files = [x for x in os.listdir(path) if x.endswith('.pickle')]
    
    master_list = []
    for name in files:
        with open(name, 'rb') as f:
            d = pickle.load(f)
        a = dict([(x,d['eval_dict'][x][metrics_map[metric]]) for x in d['eval_dict']])
        a['instance_name'] = d['instance_name']
        master_list.append(a)
        
    df = pd.DataFrame(master_list)
    df.rename(columns={'instance_name':'model'}, inplace=True)
    df.set_index('model', inplace=True)
    return df
```
